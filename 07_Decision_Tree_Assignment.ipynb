```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('50_Startups.csv')
```

```python
datasets=pd.get_dummies(dataset,drop_first=True)
```

```python
indep=datasets[['R&D Spend','Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dep=datasets[['Profit']]
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
# decision tree parameters to list variables
# https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html
criterion_list=['squared_error','friedman_mse','absolute_error','poisson']
max_features_list=[None,'sqrt','log2']
splitter_list=['best','random']
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

# list to store values
col_1=[] # my_criterion list
col_2=[] # my_max_features
col_3=[] # my_splitter 
col_4=[] # r_score 

for my_criterion in criterion_list:
    for my_max_features in max_features_list:
        for my_splitter in splitter_list:
            regressor = DecisionTreeRegressor(criterion=my_criterion, max_features=my_max_features, splitter=my_splitter )
            regressor = regressor.fit(X_train, y_train)
            y_pred = regressor.predict(X_test)
            r_score = r2_score(y_test, y_pred)
            col_1.append(my_criterion)
            col_2.append(my_max_features)
            col_3.append(my_splitter)
            col_4.append(r_score)
```

```python
# data frame header
column_names = ['CRITERION', 'MAX_FEATURES', 'SPLITTER', 'R_VALUE'] 
col_2 = [str(item) if item is None else item for item in col_2]
my_table = pd.DataFrame(list(zip(col_1,col_2,col_3,col_4)), columns=column_names)
```

```python
# output to csv file
my_table.to_csv('decision_tree_hyper_tuning_parameter.csv', index=False)
```
