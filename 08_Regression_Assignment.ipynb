```python
import pandas as pd
dataset=pd.read_csv("insurance_pre.csv")
```

```python
# Get the dimensions of the DataFrame
dimensions = dataset.shape
print(f"DataFrame dimensions (rows, columns): {dimensions}")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
independent=dataset[['age', 'bmi', 'children', 'smoker_yes']]
dependent=dataset[["charges"]]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

# Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
y_pred=regressor.predict(X_test)

from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
# Muliple Linear Reg
def mulitiple_linear_regression(X_train,y_train):
    from sklearn.linear_model import LinearRegression
    regressor=LinearRegression()
    regressor.fit(X_train,y_train)
    y_pred=regressor.predict(X_test)

    from sklearn.metrics import r2_score
    return r2_score(y_test,y_pred)   
```

```python
MLR_r_score = mulitiple_linear_regression(X_train,y_train)
```

```python
MLR_r_score
```

# Support Vector Regression ( SVR )

```python
# list to store values
col_1=[] # kernal list
col_2=[] # C value
col_3=[] # Gamma argument 
col_4=[] # r_score 

from sklearn.svm import SVR
from sklearn.metrics import r2_score

kernal_list=['linear', 'rbf','sigmoid','poly']
C_list=[10,100,1000,2000,3000]
gamma_list=['scale']

for kernal_value in kernal_list:
    for c_value in C_list:
        for gamma_value in gamma_list:
            regressor=SVR(kernel=kernal_value,C=c_value,gamma=gamma_value)
            regressor = regressor.fit(X_train, y_train)
            y_pred = regressor.predict(X_test)
            r_score = r2_score(y_test, y_pred)
            col_1.append(kernal_value)
            col_2.append(c_value)
            col_3.append(gamma_value)
            col_4.append(r_score)

# data frame header for SVM
column_names = ['kernal', 'C', 'Gamma','R_VALUE'] 
my_svm_table = pd.DataFrame(list(zip(col_1,col_2,col_3,col_4)), columns=column_names)
```

# Decision Tree

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

# decision tree parameters to list variables
criterion_list=['squared_error','friedman_mse','absolute_error','poisson']
max_features_list=[None,'sqrt','log2']
splitter_list=['best','random']

# list to store values
col_1=[] # my_criterion list
col_2=[] # my_max_features
col_3=[] # my_splitter 
col_4=[] # r_score 

for my_criterion in criterion_list:
    for my_max_features in max_features_list:
        for my_splitter in splitter_list:
            regressor = DecisionTreeRegressor(criterion=my_criterion, max_features=my_max_features, splitter=my_splitter )
            regressor = regressor.fit(X_train, y_train)
            y_pred = regressor.predict(X_test)
            r_score = r2_score(y_test, y_pred)
            col_1.append(my_criterion)
            col_2.append(my_max_features)
            col_3.append(my_splitter)
            col_4.append(r_score)

# data frame header for decision tree
column_names = ['CRITERION', 'MAX_FEATURES', 'SPLITTER', 'R_VALUE'] 
col_2 = [str(item) if item is None else item for item in col_2]
my_dt_table = pd.DataFrame(list(zip(col_1,col_2,col_3,col_4)), columns=column_names)
```

```python
my_dt_table
```

```python
# decison tree best r_values
dt_best_r = my_dt_table['R_VALUE'].max()
dt_best_r
```

# Random Forest Regressor

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html
from sklearn.ensemble import RandomForestRegressor

# random forest parameters to list variables
criterion_list=['squared_error','friedman_mse','absolute_error','poisson']
max_features_list=[None,'sqrt','log2']
n_estimate=[100]

# list to store values
col_1=[] # my_criterion list
col_2=[] # my_max_features
col_3=[] # my_splitter 
col_4=[] # r_score 

for my_criterion in criterion_list:
    for my_max_features in max_features_list:
        for my_n_estimate in n_estimate:
            regressor = RandomForestRegressor(criterion=my_criterion, max_features=my_max_features, n_estimators=my_n_estimate )
            regressor = regressor.fit(X_train, y_train)
            y_pred = regressor.predict(X_test)
            r_score = r2_score(y_test, y_pred)
            col_1.append(my_criterion)
            col_2.append(my_max_features)
            col_3.append(my_splitter)
            col_4.append(r_score)

# data frame header for decision tree
column_names = ['CRITERION', 'MAX_FEATURES', 'N_ESTIMATORS', 'R_VALUE'] 
col_2 = [str(item) if item is None else item for item in col_2]
my_rf_table = pd.DataFrame(list(zip(col_1,col_2,col_3,col_4)), columns=column_names)
```

```python
# random forest best r_values
rf_best_r = my_rf_table['R_VALUE'].max()
rf_best_r
```
